```python
import bs4 as bs
import requests
import yfinance as yf
import time
import pandas as pd
import numpy as n
import xlwings as xw
import re


url = 'https://www.slickcharts.com/sp500'
headers = {"User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
request = requests.get(url, headers = headers)
data = pd.read_html(request.text)[0]
stk_list = data.Symbol
stk_list = data.Symbol.apply(lambda x: x.replace('.', '-'))
top_15 = stk_list.head(15)
top_15
for i in top_15:
        stock = yf.Ticker(i)
        history = stock.history(period = '12mo')
        history.to_csv(i+'.csv')
        time.sleep(1)
```

```python
data = pd.read_csv('AAPL.csv')
data     
def calculate_kd(data,n=9):
    high = data['High']
    low = data['Low']
    close = data['Close']
    rsv = (close - low.rolling(window=n).min()) / (high.rolling(window=n).max() - low.rolling(window=n).min()) * 100
    k = pd.Series(0.0, index=data.index)
    d = pd.Series(0.0, index=data.index)
    
    for i in range(n, len(data)):
        k[i] = (2/3) * k[i-1] + (1/3) * rsv[i]
        d[i] = (2/3) * d[i-1] + (1/3) * k[i]
    data['K'] = k
    data['D'] = d
    return data
data = calculate_kd(data)
data.to_csv('AAPL.csv')
```

```python
def first_day(sheet):
    price = sheet.range(f"B11").value
    K = sheet.range(f"E11").value
    D = sheet.range(f"F11").value
    shares = sheet.range("L2").value
    if price > K and K<=20:
        sheet.range(f"G11").value = shares
    else:
        sheet.range(f"G11").value = 0
        
    sheet.range("I11").value = sheet.range("G11").value - sheet.range("H11").value
    sheet.range("J11").value = sheet.range("L2").value - sheet.range("I11").value * sheet.range("B11").value
    sheet.range("K11").value = sheet.range("J11").value + sheet.range("I11").value * sheet.range("B11").value
```

```python
def kd_back_test(sheet):
    Shares = sheet.range("L2").value
    last_row = sheet.range("B1").end("down").row
    
    for i in range(10, last_row + 1):
        sheet.range(f"G{i}").formula = f"=IF(AND(E{i}<=20, E{i}<F{i}, B{i}*Share<=I{i-1}), {shares}, 0)"
        sheet.range(f"H{i}").formula = f"=IF(AND(E{i}>=80, E{i}>F{i}, F{i-1}>=Share), {shares}, 0)"
        sheet.range(f"I{i}").formula = f"=G{i}-H{i}+I{i-1}"
        sheet.range(f"J{i}").formula = f"=(H{i}-G{i})*B{i}+J{i-1}"
        sheet.range(f"K{i}").formula = f"=I{i}*B{i}+J{i}"

    sheet.range("N2").value = sheet.range(f"K{last_row}").value - sheet.range(f"L2").value
```

```python
for sheet in wb.sheets:
    sheet.activate()
    kd_back_test(sheet)
    first_day(sheet)
```
