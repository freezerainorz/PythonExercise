```python
import requests
from bs4 import BeautifulSoup
rs = requests.session()
res = rs.get('https://rent.591.com.tw/?kind=1&region=1&section=5')
soup = BeautifulSoup(res.text, 'lxml')
token = soup.select_one('meta[name="csrf-token"]')['content']
print(token)
```

```python
headers = {
'X-CSRF-TOKEN': token,
'X-Requested-With': 'XMLHttpRequest' 
}

apiurl = ('https://rent.591.com.tw/home/search/rsList?is_new_list=1&type=1&kind=1&searchtype=1&region=1&section=5&firstRow=30&totalRows=852')
res = rs.get(apiurl, headers = headers)
jd = res.json()
jd
```

資料清理

```python
import pandas
df = pandas.DataFrame(jd['data']['data'])
df
```

```python
df = df[['address', 'floor', 'allfloor', 'layout', 'area', 'price', 'browsenum_all']]
df
```

```python
df.info()
```

```python
n = '80,000'
int(n.replace(',',''))
```

```python
def convert(n):
    return int(n.replace(',', ''))

n = '119,000'
convert(n)
```

```python
df['price'] = df['price'].map(lambda e: int(e.replace(',', '')))
```

```python
df.info()
```

```python
df.head()
```

```python
df[['layout_bedroom','layout_livingroom','layout_bathroom']] = df['layout'].str.extract('(\d+)房(\d+)廳(\d+)衛')
```

```python
df
```

```python
df.info()
```

```python
df['layout_bedroom'].fillna(method='ffill', inplace=True)
df['layout_livingroom'].fillna(method='ffill', inplace=True)
df['layout_bathroom'].fillna(method='ffill', inplace=True)
```

```python
df.info()
```

```python
df['layout_bedroom'] = df['layout_bedroom'].astype(int)
df['layout_livingroom'] = df['layout_livingroom'].astype(int)
df['layout_bathroom'] = df['layout_bathroom'].astype(int)
```

```python
df.head()
```

```python
#del df['layout']
df.drop('layout', axis=1,inplace=True)
```
