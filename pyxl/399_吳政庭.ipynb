```python
import bs4 as bs
import requests
import yfinance as yf
import time
import pandas as pd
import numpy as n
import xlwings as xw

url = 'https://www.slickcharts.com/sp500'
headers = {"User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
request = requests.get(url, headers = headers)
data = pd.read_html(request.text)[0]
stk_list = data.Symbol
stk_list = data.Symbol.apply(lambda x: x.replace('.', '-'))
top_15 = stk_list.head(15)
top_15
for i in top_15:
        stock = yf.Ticker(i)
        stock.history(period = '6mo').to_csv(i+'.csv')
        time.sleep(5)
    
```

```python
data = pd.read_csv('AAPL.csv')
data     
def calculate_kd(data,n=9):
    high = data['High']
    low = data['Low']
    close = data['Close']
    rsv = (close - low.rolling(window=n).min()) / (high.rolling(window=n).max() - low.rolling(window=n).min()) * 100
    k = pd.Series(0.0, index=data.index)
    d = pd.Series(0.0, index=data.index)
    
    for i in range(n, len(data)):
        k[i] = (2/3) * k[i-1] + (1/3) * rsv[i]
        d[i] = (2/3) * d[i-1] + (1/3) * k[i]
    data['K'] = k
    data['D'] = d
    return data
data = calculate_kd(data)
data.to_csv('AAPL.csv')
```

```python
stocks = ["AAPL", "AMZN", "AVGO", "BRK-B", "GOOG", "GOOGL", "JPM", "LLY", "META", "MSFT", "NVDA", "TSLA", "UNH", "V", "XOM"]
headers = ["日期", "收盤價","最高值","最低值","K值", "D值"]
new_wb = xw.Book()
for stock in stocks:
    wb = xw.Book(f"{stock}.csv")
    sheet = wb.sheets[0]
    last_row = sheet.range("B1").end("down").row
    new_sheet = new_wb.sheets.add(name = stock)
    new_sheet.range("A1").value = sheet.range(f"B1:B{last_row}").options(ndim=2).value
    new_sheet.range("B1").value = sheet.range(f"F1:F{last_row}").options(ndim=2).value
    new_sheet.range("C1").value = sheet.range(f"D1:D{last_row}").options(ndim=2).value
    new_sheet.range("D1").value = sheet.range(f"E1:E{last_row}").options(ndim=2).value
    new_sheet.range("E1").value = sheet.range(f"J1:J{last_row}").options(ndim=2).value
    new_sheet.range("F1").value = sheet.range(f"K1:K{last_row}").options(ndim=2).value
    new_sheet.range("A1").value = headers
    
new_wb.save(r"399_吳政庭.xlsx")
```
